```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates # Import for date formatting
import warnings

# Optional: Suppress FutureWarning messages for testing purposes
# warnings.filterwarnings("ignore", category=FutureWarning)

# Load the dataset
try:
    df = pd.read_csv('order_history_kaggle_data.csv')
except FileNotFoundError:
    print("Error: 'order_history_kaggle_data.csv' not found. Please ensure the file is in the correct directory.")
    exit()

# --- Graph 1: Order Status Distribution ---
order_status_counts = df['Order Status'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=order_status_counts.index, y=order_status_counts.values, palette='viridis')
plt.title('Distribution of Order Status')
plt.xlabel('Order Status')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('order_status_distribution.png')
plt.close()

# --- Graph 2: Top 10 Restaurants by Order Count ---
top_restaurants = df['Restaurant name'].value_counts().head(10)

plt.figure(figsize=(12, 7))
sns.barplot(x=top_restaurants.index, y=top_restaurants.values, palette='magma')
plt.title('Top 10 Restaurants by Order Count')
plt.xlabel('Restaurant Name')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('top_10_restaurants.png')
plt.close()

# --- Graph 3: Distribution of Total Bill Amount ---
plt.figure(figsize=(10, 6))
sns.histplot(df['Total'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Total Bill Amount')
plt.xlabel('Total Bill Amount (INR)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig('total_bill_amount_distribution.png')
plt.close()

# --- Graph 4: Orders Over Time (Daily) - FIXED CODE FOR MATPLOTLIB DEPRECATION WARNING ---
# Convert 'Order Placed At' to datetime objects with explicit format
df['Order Placed At'] = pd.to_datetime(df['Order Placed At'], format='%I:%M %p, %B %d %Y', errors='coerce')

# Drop rows where 'Order Placed At' could not be parsed (if any)
df.dropna(subset=['Order Placed At'], inplace=True)

# Extract just the date
df['Order Date'] = df['Order Placed At'].dt.date

# Get daily order counts
daily_orders = df['Order Date'].value_counts().sort_index()

# If there are gaps in dates, fill them with 0 to ensure a continuous line plot
min_date = daily_orders.index.min() if not daily_orders.empty else None
max_date = daily_orders.index.max() if not daily_orders.empty else None

if min_date and max_date:
    all_dates = pd.date_range(start=min_date, end=max_date, freq='D').date
    daily_orders = daily_orders.reindex(all_dates, fill_value=0)
else:
    daily_orders = pd.Series(dtype=int) # Empty series if no valid dates

plt.figure(figsize=(14, 7))
if not daily_orders.empty:
    # Convert dates to numbers for plotting to avoid Matplotlib deprecation warning
    x_values_numeric = mdates.date2num(daily_orders.index)
    plt.plot(x_values_numeric, daily_orders.values, marker='o', linestyle='-')

    # Format the x-axis to show dates correctly
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())
else:
    plt.text(0.5, 0.5, "No valid date data to plot", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)
plt.title('Daily Order Trends')
plt.xlabel('Date')
plt.ylabel('Number of Orders')
plt.grid(True)
plt.xticks(rotation=60, ha='right')
plt.tight_layout()
plt.savefig('daily_order_trends_resolved.png')
plt.close()

# --- Graph 5: Average Order Value by Top 10 Subzone ---
# Ensure 'Total' column is numeric. If not, coerce errors to NaN and drop.
df['Total'] = pd.to_numeric(df['Total'], errors='coerce')
df.dropna(subset=['Total'], inplace=True)

if not df.empty:
    avg_total_by_subzone = df.groupby('Subzone')['Total'].mean().sort_values(ascending=False).head(10)
else:
    avg_total_by_subzone = pd.Series(dtype=float) # Empty series if DataFrame is empty

plt.figure(figsize=(12, 7))
if not avg_total_by_subzone.empty:
    sns.barplot(x=avg_total_by_subzone.index, y=avg_total_by_subzone.values, palette='plasma')
else:
    plt.text(0.5, 0.5, "No valid data for average order value by subzone", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)
plt.title('Average Order Value by Top 10 Subzone')
plt.xlabel('Subzone')
plt.ylabel('Average Total Bill Amount (INR)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('average_order_value_by_subzone.png')
plt.close()
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
import warnings
from IPython.display import display, Image # Import for displaying images in Colab/Jupyter

# Optional: Suppress FutureWarning messages for testing purposes
# warnings.filterwarnings("ignore", category=FutureWarning)

# Load the dataset
try:
    df = pd.read_csv('order_history_kaggle_data.csv')
except FileNotFoundError:
    print("Error: 'order_history_kaggle_data.csv' not found. Please ensure the file is in the correct directory.")
    # Exit if file not found to prevent further errors
    exit()

# --- Displaying DataFrame Head as a Table ---
print("--- DataFrame Head ---")
display(df.head()) # Use display() for rich HTML table output in Colab

# --- Displaying DataFrame Info ---
print("\n--- DataFrame Info ---")
df.info() # df.info() prints directly and is usually well-formatted

print("\n--- Generating and Displaying Dashboard Graphs ---")

# --- Graph 1: Order Status Distribution ---
order_status_counts = df['Order Status'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=order_status_counts.index, y=order_status_counts.values, palette='viridis')
plt.title('Distribution of Order Status')
plt.xlabel('Order Status')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('order_status_distribution.png')
plt.close() # Close the plot to free memory
display(Image(filename='order_status_distribution.png')) # Display the saved image

# --- Graph 2: Top 10 Restaurants by Order Count ---
top_restaurants = df['Restaurant name'].value_counts().head(10)
plt.figure(figsize=(12, 7))
sns.barplot(x=top_restaurants.index, y=top_restaurants.values, palette='magma')
plt.title('Top 10 Restaurants by Order Count')
plt.xlabel('Restaurant Name')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('top_10_restaurants.png')
plt.close()
display(Image(filename='top_10_restaurants.png'))

# --- Graph 3: Distribution of Total Bill Amount ---
plt.figure(figsize=(10, 6))
sns.histplot(df['Total'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Total Bill Amount')
plt.xlabel('Total Bill Amount (INR)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig('total_bill_amount_distribution.png')
plt.close()
display(Image(filename='total_bill_amount_distribution.png'))

# --- Graph 4: Orders Over Time (Daily) - FIXED CODE FOR MATPLOTLIB DEPRECATION WARNING ---
df['Order Placed At'] = pd.to_datetime(df['Order Placed At'], format='%I:%M %p, %B %d %Y', errors='coerce')
df.dropna(subset=['Order Placed At'], inplace=True)
df['Order Date'] = df['Order Placed At'].dt.date
daily_orders = df['Order Date'].value_counts().sort_index()

min_date = daily_orders.index.min() if not daily_orders.empty else None
max_date = daily_orders.index.max() if not daily_orders.empty else None

plt.figure(figsize=(14, 7))
if min_date and max_date:
    all_dates = pd.date_range(start=min_date, end=max_date, freq='D').date
    daily_orders = daily_orders.reindex(all_dates, fill_value=0)

    # Convert dates to numbers for plotting to avoid Matplotlib deprecation warning
    x_values_numeric = mdates.date2num(daily_orders.index)
    plt.plot(x_values_numeric, daily_orders.values, marker='o', linestyle='-')

    # Format the x-axis to show dates correctly
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())
    plt.title('Daily Order Trends')
    plt.xlabel('Date')
    plt.ylabel('Number of Orders')
    plt.grid(True)
    plt.xticks(rotation=60, ha='right')
else:
    plt.text(0.5, 0.5, "No valid date data to plot", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)
    plt.title('Daily Order Trends (No data)')
    plt.xlabel('')
    plt.ylabel('')
plt.tight_layout()
plt.savefig('daily_order_trends_resolved.png')
plt.close()
display(Image(filename='daily_order_trends_resolved.png'))


# --- Graph 5: Average Order Value by Top 10 Subzone ---
df['Total'] = pd.to_numeric(df['Total'], errors='coerce')
df.dropna(subset=['Total'], inplace=True)

plt.figure(figsize=(12, 7))
if not df.empty: # Check if df is empty after dropping NaNs
    avg_total_by_subzone = df.groupby('Subzone')['Total'].mean().sort_values(ascending=False).head(10)
    if not avg_total_by_subzone.empty:
        sns.barplot(x=avg_total_by_subzone.index, y=avg_total_by_subzone.values, palette='plasma')
        plt.title('Average Order Value by Top 10 Subzone')
        plt.xlabel('Subzone')
        plt.ylabel('Average Total Bill Amount (INR)')
        plt.xticks(rotation=45, ha='right')
    else:
        plt.text(0.5, 0.5, "No valid data for average order value by subzone", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)
        plt.title('Average Order Value by Top 10 Subzone (No data)')
        plt.xlabel('')
        plt.ylabel('')
else:
    plt.text(0.5, 0.5, "No valid data for average order value by subzone", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)
    plt.title('Average Order Value by Top 10 Subzone (No data)')
    plt.xlabel('')
    plt.ylabel('')
plt.tight_layout()
plt.savefig('average_order_value_by_subzone.png')
plt.close()
display(Image(filename='average_order_value_by_subzone.png'))

print("\nAll dashboard graphs should now be displayed in the output cells above and saved as PNG files.")
print("Remember, the FutureWarning about DataFrame.count() is a notification for future Pandas versions and does not prevent code execution or graph generation.")
print("All dashboard graphs have been generated and saved as PNG files in the current directory.")
```
